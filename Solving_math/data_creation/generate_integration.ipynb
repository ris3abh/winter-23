```python
import numpy as np
import random
import math
import sympy as sp
import pandas as pd
import os
import sys
import time
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sympy import *
from sympy.parsing.sympy_parser import parse_expr
from sympy.parsing.sympy_parser import standard_transformations, implicit_multiplication_application
```

```python
def integrate(equation, x):
    integration = sp.integrate(equation, x)
    return integration
```

```python
## generating equations for integration of the form a*x + b where a and b are random integers between -1000 and 1000
def generate_integration_equation():
    ## generate random integers between -1000 and 1000 for a and b
    a = random.randint(-1000, 1000)
    b = random.randint(-1000, 1000)
    ## create the equation
    equation = a*x + b
    ## integrate the equation
    solution = integrate(equation, x)
    return equation, solution
```

```python
## testing generate_equation_integration function
## running a for loop to generate 400 equations and solutions and saving them to a csv file
equations = []
solutions = []
for i in range(400):
    equation, solution = generate_integration_equation()
    equations.append(equation)
    solutions.append(solution)
    
## saving the equations and solutions to a csv file
df = pd.DataFrame({'equation': equations, 'solution': solutions})
df.to_csv('integration1.csv', index=False)
```
