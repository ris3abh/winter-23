```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open(r'data/alice.txt', 'r') as f:
    text = f.read()
```

```python
doc = nlp(text)
```

```python
print(doc)
```

```python
sentence = list(doc.sents)
```

```python
sent = sentence[0]
```

```python
for token in sent:
    print(token.text, token.pos_)
```

```python
for token in sent:
    if token.pos_ == 'NOUN':
        print(token.text)
```
