```python
import spacy as sc
```

```python
nlp = sc.load('en_core_web_sm')
```

```python
with open(r'/Users/rishabhsharma/Documents/GitHub/winter-23/DSCI-521/assignment1/module-B/data/books/84.txt', 'r') as file:
    text = file.read().replace('\n', ' ').replace('\n\n', ' ')
```

```python
## splitting the text into half
text1 = text[:len(text)//2]
```

```python
doc = nlp(text1)
```

## Breaking down text by sentences

```python
sentences = list(doc.sents)
```

```python
sentences[10]
```

## Named Entity Recognition (NER)

```python
## creating ents(entities)

sent = sentences[33]
print(sent)
ents = list(sent.ents)
```

```python
for i in ents:
    print(i)
```

```python
ents = list(doc.ents)
```

```python
for i in ents:
    print(i.text, i.label_, i.label)
```

```python
dict = {}
for i in ents:
    if i.label_ in dict:
        dict[i.label_] += 1
    else:
        dict[i.label_] = 1
```
